## ANALYSE EN COMPOSANTE PRINCIPALE

```python
# Création d'un tableau 2D à partir du dataframe data
df = data.values

# Création d'un pipeline pour la standardisation et l'ACP
pipe = Pipeline([('standardisation', StandardScaler()), ('ACP', PCA(n_components = 0.95))])

# Application de pipe aux dataset
pipe.fit(df)

# Visualisation du pourcentage de variance expliquée par chaque composante
plt.figure(figsize = (15,8))
plt.bar(range(pipe[1].n_components_), pipe[1].explained_variance_ratio_)
plt.xlabel('composantes principales')
plt.ylabel('pourcentage de variance')
plt.xticks(range(pipe[1].n_components_))
plt.show()
```

Avec les moyennes mobiles nous recherchons une tendance dans la cible. 

```python
# Affichage des moyenne mobile
plt.figure(figsize = (15, 8))
orig = plt.plot(df['Adj Close'], color = 'blue', label = 'Original')
moy = plt.plot(mobile, color = 'red', label = 'Rolling mean')
std = plt.plot(mobilestd, color = 'black', label = 'rolling std')
plt.legend(loc = 'best')
plt.title('roll et std deviation')
plt.show(block = False)
```

Nous voyons à travers la courbe de moyenne mobile que la courbe ne suit pas de tendance constante. elle n'a pas l'air de fluctuer autour d'une valeur. Elle n'est donc pas stationnaire. Pour en être su nous pouvons faire le test de stationnarité de Dickey Fuller

```python
print('Resultat du test de dickey fuller: ')
DFtest = adfuller(df['Adj Close'], autolag = 'AIC')
dfoutput = pd.Series(DFtest[0:4], index = ['Test Statistic', 'p-value', 'lags used','nbre observation utilisé'])
for cle,valeur in DFtest[4].items():
    dfoutput['valeur critique (%s)'%cle] = valeur
    
print(dfoutput)
```

La p-value est supérieur au seuil donc la série est non stationnaire.
